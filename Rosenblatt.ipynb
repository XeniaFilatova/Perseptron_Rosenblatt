```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron():

    def __init__(self, inputs=2):
        self.weights = 2 * np.random.rand(inputs).T - 1 # Веса, -1 > w < 1
        self.b = np.random.rand(1) - 1
        print("start data: {}, b: {}".format(self.weights, self.b))
        
    def predict(self, x):
        y = 1 if (np.dot(x, self.weights) + self.b >= 0) else 0
        return y

    def train(self, X, Y, epochs=2):
        self.show_plot(X, Y, 0)
        for i in range(epochs):
            for x_curr, y_curr in zip(X,Y):
                y_predict = self.predict(x_curr)
                err = ( -y_predict + y_curr)
                self.weights += err * x_curr
                self.b += err
                print("x: {}, y: {}, y_predict: {}, mae {}".format(x_curr, y_curr, y_predict, err))
                print("weight data: {}, b: {}".format(self.weights, self.b))
            self.show_plot(X, Y, i)
            

    def show_plot(self, X, Y, i):
        print("Epoch: #{}, weights: {}, b: {}".format(i+1, self.weights, self.b))
        x1 = X[Y==0][:,0]
        y1 = X[Y==0][:,1]
        x2 = X[Y==1][:,0]
        y2 = X[Y==1][:,1]
        plt.figure(figsize=(7,4))
        x_line = np.arange(-5, 5, 0.1)
        y_line = np.apply_along_axis(lambda t: (-self.b - t * self.weights[0])/self.weights[1], 0, x_line)
        plt.scatter(x1,y1, color = "red")
        plt.scatter(x2,y2, color = "green")
        plt.plot(x_line, y_line, color = "blue")
        plt.show()
```

```python
X = np.array([[3, 2.4], 
              [-3.8, 0.2], 
              [-1.8, 0.4], 
              [-1.1, -0.9], 
              [-3.2, -2.5],
              [-4.8, 4.2]])
               
Y = np.array([0, 1, 1, 1, 1, 0])


perceptron = Perceptron(inputs=2)

perceptron.train(X, Y)


```
